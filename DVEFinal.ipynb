```python
#function to plot dates 
def plotDate (index1,index2,g):
    data=g[index1:index2]
    x=data.index #index are the dates since we grouped by date 
    x=x.strftime('%Y-%m-%d') #converts date to string
    yvalues=data.values # values are the number of users per day 

    mytitle="Users dating from "+x[0]+" until "+x[len(x)-1]
   
    df = pd.DataFrame({'Date':x, 'No of unique users':yvalues})
    ax = df.plot.bar(x='Date', y='No of unique users',rot=0,title=mytitle,figsize=(20,5))
    ax.set_xticklabels(labels=x, rotation=70, rotation_mode="anchor", ha="right")
```

6. For each Tuesday and Wednesday, an indicator will added for whether the price goes up or the price goes
down that month.

    0: If the price goes down 

    1: If the price goes up

    The first entry will hold 1 

```python
#For Tuesday 
TueDf['dfIndex'] = np.array(range(len(TueDf.index))) #creates a column that holds the index (int) of the row 

TueDf['PriceUp/Down']=0 #sets everything to 0

for i in range(0,len(TueDf.index)):
    if(i==0):
        TueDf['PriceUp/Down']=TueDf.apply(lambda x: 1 if x['dfIndex']==i else x['PriceUp/Down'],axis=1) #sets first entry to 1
    if(TueDf.iloc[i-1]['cost_per_gallon_float']<TueDf.iloc[i]['cost_per_gallon_float']): #check previous cost againts current cost 
        TueDf['PriceUp/Down']=TueDf.apply(lambda x: 1 if x['dfIndex']==i else x['PriceUp/Down'],axis=1) #if price goes up, assigns a 1
    
print(TueDf[['date_fueled','cost_per_gallon_float','PriceUp/Down']])


#For Wednesay
WedDf['dfIndex'] = np.array(range(len(WedDf.index))) #creates a column that holds the index (int) of the row 

WedDf['PriceUp/Down']=0 #sets everything to 0

for i in range(0,len(WedDf.index)):
    if(i==0):
        WedDf['PriceUp/Down']=WedDf.apply(lambda x: 1 if x['dfIndex']==i else x['PriceUp/Down'],axis=1) #sets first entry to 1
    if(WedDf.iloc[i-1]['cost_per_gallon_float']<WedDf.iloc[i]['cost_per_gallon_float']): #check previous cost againts current cost 
        WedDf['PriceUp/Down']=WedDf.apply(lambda x: 1 if x['dfIndex']==i else x['PriceUp/Down'],axis=1) #if price goes up, assigns a 1
    
print(WedDf[['date_fueled','cost_per_gallon_float','PriceUp/Down']])

```

7. Do more people refuel on the first Wednesday of the month when the prices goes down?
